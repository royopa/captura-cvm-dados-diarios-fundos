```python
import pandas as pd
import dask.dataframe as dd
```

```python
df = dd.read_csv('saida.csv')
```

```python
print('Convertendo o df dask para pandas')
df = df.compute()
```

```python
# faz o sort do df
df.sort_values(by=['CO_PRD', 'DT_REF'], inplace=True)
```

```python
df['CO_PRD'] = df['CO_PRD'].astype(str).str.zfill(14)
df['DT_REF'] = pd.to_datetime(df['DT_REF'], format='%Y-%m-%d', errors='coerce')
```
